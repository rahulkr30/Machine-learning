```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler

# Load the Diabetes dataset
diabetes = load_diabetes()
X, y = diabetes.data, diabetes.target
feature_names = diabetes.feature_names

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define a range of alpha values
alpha_values = [0.1, 1, 10, 100,1000,10000]

# Create subplots for the bar graphs
fig, axes = plt.subplots(len(alpha_values), figsize=(10, 8))

# Perform ridge regression and plot the coefficient magnitudes for each alpha
for i, alpha in enumerate(alpha_values):
    ridge = Ridge(alpha=alpha)
    ridge.fit(X_scaled, y)
    coef_magnitudes = np.abs(ridge.coef_)
    axes[i].bar(feature_names, coef_magnitudes)
    axes[i].set_title(f"Alpha = {alpha}")
    axes[i].set_ylabel("Coefficient Magnitude")
    axes[i].tick_params(axis="x", rotation=90)

plt.tight_layout()
plt.show()
```

# impact of lamda in loss function and coficient 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error 
from sklearn.datasets import load_diabetes
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler

# Load the diabetes dataset
diabetes = load_diabetes()
X, y = diabetes.data, diabetes.target

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define a range of lambda values
lambda_values = [0.001, 0.01, 0.1, 1, 10, 100]

# Perform ridge regression for each lambda value
coefficients = []
losses = []

for lambda_val in lambda_values:
    ridge = Ridge(alpha=lambda_val)
    ridge.fit(X_scaled, y)
    
    coefficients.append(ridge.coef_)
    
    y_pred = ridge.predict(X_scaled)
    loss = mean_squared_error(y, y_pred)
    losses.append(loss)

# Plot the coefficient estimates
plt.figure(figsize=(12, 6))
for i, feature in enumerate(diabetes.feature_names):
    plt.plot(lambda_values, [coef[i] for coef in coefficients], label=feature)
plt.xscale('log')
plt.xlabel('Lambda')
plt.ylabel('Coefficient Estimate')
plt.title('Impact of Lambda on Coefficient Estimates')
plt.legend()
plt.show()

# Plot the loss function
plt.figure(figsize=(8, 6))
plt.plot(lambda_values, losses)
plt.xscale('log')
plt.xlabel('Lambda')
plt.ylabel('Mean Squared Error')
plt.title('Impact of Lambda on Loss Function')
plt.show()
```

```python
# Define a range of lambda values
lambda_values = [0.001, 0.01, 0.1, 1, 10, 100]

# Compute the inverse of XX for each lambda value
inverses = []

for lambda_val in lambda_values:
    Xtx = np.dot(X_scaled.T, X_scaled)
    Xtx_reg = Xtx + lambda_val * np.identity(Xtx.shape[0])
    Xtx_inv = np.linalg.inv(Xtx_reg)
    inverses.append(Xtx_inv)

# Plot the inverse of XX
plt.figure(figsize=(8, 6))
for i in range(X.shape[1]):
    plt.plot(lambda_values, [inv[i, i] for inv in inverses], label=f'Column {i+1}')
plt.xscale('log')
plt.xlabel('Lambda')
plt.ylabel('Value')
plt.title('Impact of Lambda on Inverse of XX')
plt.legend()
plt.show()
```
